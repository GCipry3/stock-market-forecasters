# IMPORTS

```python
import yfinance as yf
import pandas as pd
import numpy as np
from darts import TimeSeries
from darts.models import RegressionModel
from darts.metrics import mae, mse
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt
```

# UTILS

```python
def load_data(ticker="^GSPC", start_date='1990-01-02', end_date='2024-03-20', split_date='2023-01-01'):
    df = yf.download(ticker, start=start_date, end=end_date)
    df['Price'] = df['Close']
    df = df[['Price']]
    
    business_days = pd.date_range(start=start_date, end=end_date, freq='B')
    df = df.reindex(business_days).ffill()

    # Convert the DataFrame into a TimeSeries data type
    price_ts = TimeSeries.from_dataframe(df, value_cols=['Price'], fill_missing_dates=True, freq='B')
    
    # Split the data into train and test TimeSeries
    price_train_ts, price_test_ts = price_ts.split_before(pd.Timestamp(split_date))
    return price_train_ts, price_test_ts

def train_polynomial_regression(train_ts, degree=2):
    # Extract the time series into numpy arrays for the regression
    y_train = train_ts.values().flatten()
    x_train = np.arange(len(y_train)).reshape(-1, 1)
    
    # Create a polynomial regression model with specified degree
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(x_train, y_train)
    return model

def plot_predictions(test_ts, predicted_ts, title=""):
    plt.figure(figsize=(12, 6))
    test_ts.plot(label='Actual')
    predicted_ts.plot(label='Forecast')
    plt.title(title)
    plt.legend()
    plt.show()

def plot_full_series(train_ts, test_ts, predicted_ts, title=""):
    plt.figure(figsize=(15, 8))
    train_ts.plot(label='Training Data')
    test_ts.plot(label='Test Data', color='orange')
    predicted_ts.plot(label='Forecast', color='red')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
price_train_ts, price_test_ts = load_data(split_date="2020-01-01")
model = train_polynomial_regression(price_train_ts, degree=5) 
predictions_length = len(price_test_ts)
predicted_values = model.predict(np.arange(len(price_train_ts) + predictions_length).reshape(-1, 1))[-predictions_length:]
predicted_ts = TimeSeries.from_times_and_values(price_test_ts.time_index, predicted_values)

print(f"MSE on test set: {mse(price_test_ts, predicted_ts)}")
print(f"MAE on test set: {mae(price_test_ts, predicted_ts)}")

plot_predictions(price_test_ts, predicted_ts, title="Polynomial Regression Forecast of SP500 Prices")
plot_full_series(price_train_ts, price_test_ts, predicted_ts, title="Polynomial Regression Forecast of SP500 Prices")
```
